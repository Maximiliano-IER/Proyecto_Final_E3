```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse
```

```python
def importa(ArchivoExp):
    DataExp = pd.read_csv(ArchivoExp,sep=',',names=nombres2, infer_datetime_format=True,parse_dates=[0])
    DataExp.set_index('tiempo',inplace=True)
    DataExp['TCPan'] = (DataExp.TCPan1 + DataExp.TCPan2 + DataExp.TCPan3 + DataExp.TCPan4 + DataExp.TCPan5 + DataExp.A1 + DataExp.A2 + DataExp.A3 + DataExp.A4)/9
    return DataExp
nombres2 = ['tiempo' ,'SFlujo'   , 'TemPrtdIn', 'TemPrtdOut','TCPan1'  , 'TCPan2'   , 'TCPan3'      , 'TCPan4'  , 'TCPan5',
           'TCCubo01', 'TCCubo02', 'TCCubo03' , 'TCCubo04'  ,'TCCubo05', 'TCCubo06' , 'TCCubo07'    , 'TCCubo08', 'TCCubo09', 'TCCubo010', 
           'TCPiso'  , 'Ti'  , 'A1' , 'A2', 'A3', 'A4'  ,'Flujo2'  , 'TemPrtdIn2', 'TemPrtdOut2','InterIn'  , 'InterOut' ]
```

```python
Datos_ex = importa('../../proyecto_final_EE-main/002_datos/mediciones.csv')
Datos_ex
```

```python
Ti = Datos_ex.Ti.resample('10Min').bfill()
Ti
```

```python
CB = pd.read_csv('../IDF/Caso_Base/Proyecto_fin.csv',names=['tiempo','Ti_C1','Ti_C2','Ti_CB','Ti_techo1','Ti_techo2','Ti_techoJorge'],skiprows=1)
CB.tiempo = CB.tiempo.str.replace('24:00:00','23:59:59')
CB.tiempo = '2019 ' + CB.tiempo
CB.tiempo = pd.to_datetime(CB.tiempo,format='%Y %m/%d %H:%M:%S')
CB.set_index('tiempo',inplace=True)
del CB['Ti_C1']
del CB['Ti_C2']
del CB['Ti_techo1']
del CB['Ti_techo2']
del CB['Ti_techoJorge']
CB
```

```python
CA = pd.read_csv('../IDF/Caso_absortancia/Proyecto_fin_absortanciaa.csv',names=['tiempo','Ti_C1','Ti_C2','Ti_CA','Ti_techo1','Ti_techo2','Ti_techoJorge'],skiprows=1)
CA.tiempo = CA.tiempo.str.replace('24:00:00','23:59:59')
CA.tiempo = '2019 ' + CA.tiempo
CA.tiempo = pd.to_datetime(CA.tiempo,format='%Y %m/%d %H:%M:%S')
CA.set_index('tiempo',inplace=True)
del CA['Ti_C1']
del CA['Ti_C2']
del CA['Ti_techo1']
del CA['Ti_techo2']
del CA['Ti_techoJorge']
CA
```

```python
compara = pd.concat([CB,CA],axis=1)
compara
```

```python
CI01 = pd.read_csv('../IDF/Caso_infiltracion_01/Proyecto_CasoCambioInfiltracion.csv',names=['tiempo','Ti_C1','Ti_C2','Ti_CI01','Ti_techo1','Ti_techo2','Ti_techoJorge'],skiprows=1)
CI01.tiempo = CI01.tiempo.str.replace('24:00:00','23:59:59')
CI01.tiempo = '2019 ' + CI01.tiempo
CI01.tiempo = pd.to_datetime(CI01.tiempo,format='%Y %m/%d %H:%M:%S')
CI01.set_index('tiempo',inplace=True)
del CI01['Ti_C1']
del CI01['Ti_C2']
del CI01['Ti_techo1']
del CI01['Ti_techo2']
del CI01['Ti_techoJorge']
CI01
```

```python
compara2 = pd.concat([compara,CI01],axis=1)
compara2
```

```python
CI03 = pd.read_csv('../IDF/Caso_infiltracion_003/Proyecto_CasoCambioInfiltracion003.csv',names=['tiempo','Ti_C1','Ti_C2','Ti_CI03','Ti_techo1','Ti_techo2','Ti_techoJorge'],skiprows=1)
CI03.tiempo = CI03.tiempo.str.replace('24:00:00','23:59:59')
CI03.tiempo = '2019 ' + CI03.tiempo
CI03.tiempo = pd.to_datetime(CI03.tiempo,format='%Y %m/%d %H:%M:%S')
CI03.set_index('tiempo',inplace=True)
del CI03['Ti_C1']
del CI03['Ti_C2']
del CI03['Ti_techo1']
del CI03['Ti_techo2']
del CI03['Ti_techoJorge']
CI03
```

```python
compara3 = pd.concat([compara2,CI03],axis=1)
compara3
```

```python
Ccomb = pd.read_csv('../IDF/Caso_infil_abs/Proyecto_caso_absortancia_mas_infilt.csv',names=['tiempo','To','Ti_C1','Ti_C2','Ti_Ccomb','Ti_techo1','Ti_techo2','Ti_techoJorge'],skiprows=1)
Ccomb.tiempo = Ccomb.tiempo.str.replace('24:00:00','23:59:59')
Ccomb.tiempo = '2019 ' + Ccomb.tiempo
Ccomb.tiempo = pd.to_datetime(Ccomb.tiempo,format='%Y %m/%d %H:%M:%S')
Ccomb.set_index('tiempo',inplace=True)
del Ccomb['Ti_C1']
del Ccomb['Ti_C2']
del Ccomb['Ti_techo1']
del Ccomb['Ti_techo2']
del Ccomb['Ti_techoJorge']
Ccomb
```

```python
compara4 = pd.concat([compara3,Ccomb],axis=1)
compara4
```

```python
compara5 = pd.concat([compara4,Ti],axis=1)
compara5
```

## Grafica de datos juntos

```python
fig, ax = plt.subplots(figsize=(16,7))

fecha1 = parse("2019-06-22")
fecha2 = fecha1 + pd.Timedelta("2D")

ax.set_xlim(fecha1,fecha2)


ax.plot(compara5.Ti_ex,'b-',linewidth=4,label='Ti experimental')
ax.plot(compara5.Ti_CB,'g+',label='Ti Caso Base')
ax.plot(compara5.Ti_CA,'r+',label='Ti Caso Absortancia')
ax.plot(compara5.Ti_CI01,'y+',label='Ti Caso coef = 0.1')
ax.plot(compara5.Ti_CI03,'b+',label='Ti Caso coef = 0.03')
ax.plot(compara5.Ti_Ccomb,'r+',label='Ti Caso Infiltración + Absortancia')

ax.plot(compara5.To,'k-',linewidth=4,label='To')

#Cositas pa' darle formato
ax.set(xlabel='Tiempo', ylabel='Temperatura [°C]',
       title='Gráfica de temperatura al interior de cubículo del Dr. Jorge (2019-06-22)')
ax.legend()
ax.grid()
```
